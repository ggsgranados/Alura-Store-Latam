# 2. Ventas por categoría

```python
# Cuenta la frecuencia de cada categoría en la columna 'Categoría del Producto'
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()

# Imprime las categorías y sus conteos
print("Ventas por categorías de la Tienda 1:")
print(ventas_por_categoria_tienda1)
```

```python
# Cuenta la frecuencia de cada categoría en la columna 'Categoría del Producto'
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()

# Imprime las categorías y sus conteos
print("Ventas por categorías de la Tienda 2:")
print(ventas_por_categoria_tienda2)
```

```python
# Cuenta la frecuencia de cada categoría en la columna 'Categoría del Producto'
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()

# Imprime las categorías y sus conteos
print("Ventas por categorías de la Tienda 3:")
print(ventas_por_categoria_tienda3)
```

```python
# Cuenta la frecuencia de cada categoría en la columna 'Categoría del Producto'
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Imprime las categorías y sus conteos
print("Ventas por categorías de la Tienda 4:")
print(ventas_por_categoria_tienda4)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Obtener las categorías únicas y ordenarlas
categorias_tienda1 = ventas_por_categoria_tienda1.index.tolist()
categorias_tienda2 = ventas_por_categoria_tienda2.index.tolist()
categorias_tienda3 = ventas_por_categoria_tienda3.index.tolist()
categorias_tienda4 = ventas_por_categoria_tienda4.index.tolist()
categorias_unicas = sorted(list(set(categorias_tienda1 + categorias_tienda2 + categorias_tienda3 + categorias_tienda4)))

# Asegurar que todas las series tengan todas las categorías (llenando con 0 si falta)
ventas_tienda1_completo = ventas_por_categoria_tienda1.reindex(categorias_unicas, fill_value=0).tolist()
ventas_tienda2_completo = ventas_por_categoria_tienda2.reindex(categorias_unicas, fill_value=0).tolist()
ventas_tienda3_completo = ventas_por_categoria_tienda3.reindex(categorias_unicas, fill_value=0).tolist()
ventas_tienda4_completo = ventas_por_categoria_tienda4.reindex(categorias_unicas, fill_value=0).tolist()

# Crear el gráfico de radar
labels = np.array(categorias_unicas)
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()

# Cerrar el círculo
angles += angles[:1]
ventas_tienda1_completo += ventas_tienda1_completo[:1]
ventas_tienda2_completo += ventas_tienda2_completo[:1]
ventas_tienda3_completo += ventas_tienda3_completo[:1]
ventas_tienda4_completo += ventas_tienda4_completo[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

ax.plot(angles, ventas_tienda1_completo, linewidth=2, label='Tienda 1')
ax.plot(angles, ventas_tienda2_completo, linewidth=2, label='Tienda 2')
ax.plot(angles, ventas_tienda3_completo, linewidth=2, label='Tienda 3')
ax.plot(angles, ventas_tienda4_completo, linewidth=2, label='Tienda 4')

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Etiquetas de las categorías
ax.set_thetagrids(np.degrees(angles[:-1]), labels)

# Ajustar los límites del gráfico
max_venta = max(max(ventas_tienda1_completo), max(ventas_tienda2_completo), max(ventas_tienda3_completo), max(ventas_tienda4_completo))
ax.set_ylim(0, max_venta * 1.2)

# Título del gráfico
plt.title('Comparación de Frecuencia de Productos por Categoría (Todas las Tiendas)', pad=40)

# Leyenda
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos

```python
"""
Función para encontrar el producto más vendido y el menos vendido en cada tienda
@param df: DataFrame de la tienda
@param tienda_num: Número de la tienda
"""
def best_selling_products(df, tienda_num):
    # Agrupa por nombre de producto y suma las cantidades vendidas
    product_sales = df['Producto'].value_counts()
    # Ordena las ventas en orden descendente
    top_product = product_sales.sort_values(ascending=False)
    print(f"Tienda {tienda_num}:")
    print(f"Total de productos encontrados: {top_product.count()}")
    print("-" * 20)
    print("Top 5 productos más vendidos:\n")
    # Muestra los 5 productos más vendidos y sus cantidades
    for i in range(min(5, len(top_product))):
        print(f"Producto {i+1}:")
        print(f"{top_product.index[i]}\nCantidad vendida: {top_product.values[i]}\n")
    print("-" * 20)
    print("Top 5 productos menos vendidos:\n")
    # Muestra los 5 productos menos vendidos y sus cantidades
    for i in range(min(5, len(top_product))):
        print(f"Producto {i+1}:")
        print(f"{top_product.index[-(i+1)]}\nCantidad vendida: {top_product.values[-(i+1)]}\n")
    print("-" * 20)



```

```python
import matplotlib.pyplot as plt

"""Función para graficar los productos más y menos vendidos en cada tienda
@param df: DataFrame de la tienda
@param tienda_num: Número de la tienda
"""

def plot_product_sales(df, tienda_num):
    product_sales = df['Producto'].value_counts()
    top_products = product_sales.head(5)   # Top 5 productos mas vendidos
    bottom_products = product_sales.tail(5) # Top 5 productos menos vendidos

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    top_products.plot(kind='barh', color='skyblue')
    plt.title(f'Top 5 Productos mas vendidos - Tienda {tienda_num}')
    plt.xlabel('Cantidad vendida')
    plt.ylabel('')
    plt.yticks(rotation=0)
    plt.gca().invert_yaxis()


    plt.subplot(1, 2, 2)
    bottom_products.plot(kind='barh', color='lightcoral')
    plt.title(f'Top 5 Productos menos vendidos - Tienda {tienda_num}')
    plt.xlabel('Cantidad vendida')
    plt.ylabel('')
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.show()
```

```python
print('Productos más y menos vendidos de la Tienda 1\n')
best_selling_products(tienda, 1)
# Graficando
plot_product_sales(tienda, 1)
```

```python
print('Productos más y menos vendidos de la Tienda 2\n')
best_selling_products(tienda2, 2)
# Graficando
plot_product_sales(tienda2, 2)
```

```python
print('Productos más y menos vendidos de la Tienda 3\n')
best_selling_products(tienda3, 3)
# Graficando
plot_product_sales(tienda3, 3)
```

```python
print('Productos más y menos vendidos de la Tienda 4\n')
best_selling_products(tienda4, 4)
# Graficando
plot_product_sales(tienda4, 4)
```
