# 4. Productos más y menos vendidos

```python
"""
Función para encontrar el producto más vendido y el menos vendido en cada tienda
@param df: DataFrame de la tienda
@param tienda_num: Número de la tienda
"""
def best_selling_products(df, tienda_num):
    # Agrupa por nombre de producto y suma las cantidades vendidas
    product_sales = df['Producto'].value_counts()
    # Ordena las ventas en orden descendente
    top_product = product_sales.sort_values(ascending=False)
    print(f"Tienda {tienda_num}:")
    print(f"Total de productos encontrados: {top_product.count()}")
    print("-" * 20)
    print("Top 5 productos más vendidos:\n")
    # Muestra los 5 productos más vendidos y sus cantidades
    for i in range(min(5, len(top_product))):
        print(f"Producto {i+1}:")
        print(f"{top_product.index[i]}\nCantidad vendida: {top_product.values[i]}\n")
    print("-" * 20)
    print("Top 5 productos menos vendidos:\n")
    # Muestra los 5 productos menos vendidos y sus cantidades
    for i in range(min(5, len(top_product))):
        print(f"Producto {i+1}:")
        print(f"{top_product.index[-(i+1)]}\nCantidad vendida: {top_product.values[-(i+1)]}\n")
    print("-" * 20)



```

```python
import matplotlib.pyplot as plt

"""Función para graficar los productos más y menos vendidos en cada tienda
@param df: DataFrame de la tienda
@param tienda_num: Número de la tienda
"""

def plot_product_sales(df, tienda_num):
    product_sales = df['Producto'].value_counts()
    top_products = product_sales.head(5)  # Top 5 productos mas vendidos
    bottom_products = product_sales.tail(5)  # Top 5 productos menos vendidos

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    top_products.plot(kind='bar', color='skyblue')
    plt.title(f'Top 5 Productos mas vendidos - Tienda {tienda_num}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.xticks(rotation=45, ha='right')

    plt.subplot(1, 2, 2)
    bottom_products.plot(kind='bar', color='lightcoral')
    plt.title(f'Top 5 Productos menos vendidos - Tienda {tienda_num}')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad vendida')
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()
```

```python
print('Productos más y menos vendidos de la Tienda 1\n')
best_selling_products(tienda, 1)
# Graficando
plot_product_sales(tienda, 1)
```

```python
print('Productos más y menos vendidos de la Tienda 2\n')
best_selling_products(tienda2, 2)
# Graficando
plot_product_sales(tienda2, 2)
```

```python
print('Productos más y menos vendidos de la Tienda 3\n')
best_selling_products(tienda3, 3)
# Graficando
plot_product_sales(tienda3, 3)
```

```python
print('Productos más y menos vendidos de la Tienda 4\n')
best_selling_products(tienda4, 4)
# Graficando
plot_product_sales(tienda4, 4)
```
